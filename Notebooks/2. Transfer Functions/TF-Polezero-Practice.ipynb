## Create the and plot the step response for the following transfer functions:
 
  $$  \frac{s+1}{10s+1}$$

  $$  \frac{s+3}{7s^2 + 2s +1}$$
 
  $$  \frac{s+3}{7s^2 - 2s +1}$$
 
  $$  \frac{s+3}{3s^2 + 1}$$
  
  $$  \frac{s+3}{s^2+ 2s}$$
    
  $$  \frac{s+3}{3s^2+3s+ 1}$$
      
  
  ### Write whether the transfer fucntions are stable/unstable and if they are oscillatory/non-oscillatory

```python
import control 
import numpy as np 
import matplotlib.pyplot as plt
% matplotlib inline 
```

```python

g1 = control.tf([4,1],[10,1])
g2 = control.tf([1,3],[30,2,1])
g3 = control.tf([1,3],[30,-2,1])
g4 = control.tf([1,3],[3,0,1])
g5 = control.tf([1,3],[1,2,0])
g6 = control.tf([1,3],[3,3,1])

t = np.linspace(0,100,100)
u  =  np.zeros_like(t)
u[10:] = 1
resp1= []
t, resp1, __ = control.forced_response(g1,t,u)
t, resp2, __ = control.forced_response(g2,t,u)
t, resp3, __ = control.forced_response(g3,t,u)
t, resp4, __ = control.forced_response(g4,t,u)
t, resp5, __ = control.forced_response(g5,t,u)
t, resp6, __ = control.forced_response(g6,t,u)

plt.subplot(321,xlabel="time,t",ylabel="output")
plt.plot(t,resp1)

plt.subplot(322,xlabel="time,t",ylabel="output")
plt.plot(t,resp2)

plt.subplot(323,xlabel="time,t",ylabel="output")
plt.plot(t,resp3)

plt.subplot(324,xlabel="time,t",ylabel="output")
plt.plot(t,resp4)

plt.subplot(325,xlabel="time,t",ylabel="output")
plt.plot(t,resp5)

plt.subplot(326,xlabel="time,t",ylabel="output")
plt.plot(t,resp6)
plt.tight_layout()
```

### In Order (S/US) and (O/NO):

 1. SNO
 2. SO
 3. USO
 4. SO (marginal case)
 5. USNO
 6. SNO


### For each of the transfer functions in 1, List the Poles of the transfer functions 


```python
transfer_functions = [g1,g2,g3,g4,g5,g6]
for tf in transfer_functions:
    poles = control.pole(tf)
    print(" transfer function {} has poles at these points {}. \n".format(1+transfer_functions.index(tf),poles))
```

### List the zeros of each of the transfer functions:


```python
transfer_functions = [g1,g2,g3,g4,g5,g6]
for tf in transfer_functions:
    poles = control.zero(tf)
    print(" transfer function {} has zeros at these points {}. \n".format(1+transfer_functions.index(tf),poles))
```
